```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import auc
```

```python
roc_files = [
    '/content/drive/MyDrive/ethereum-fraud-detection/results/roc_data_logistic_reg.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/roc_data_svm.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/roc_data_neural_net.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/roc_data_random_forest.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/roc_data_xgboost.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/roc_data_lightgbm.csv',
]

labels = [
    'Logistic Regression',
    'SVM',
    'MLP',
    'Random Forest',
    'XGBoost',
    'LightGBM',
]

colors = ['cyan', 'brown', 'olive', 'magenta', 'blue', 'green']

plt.figure(figsize=(8, 6))

for i, file in enumerate(roc_files):
    roc_data = pd.read_csv(file)
    fpr = roc_data['fpr']
    tpr = roc_data['tpr']
    auc_value = auc(fpr, tpr)

    plt.plot(fpr, tpr, color=colors[i], label=f'{labels[i]} (AUC={auc_value:.2f})', linewidth=1.0)

plt.plot([0, 1], [0, 1], color='red', linestyle='--', linewidth=1.0)

plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
# plt.grid(True)
plt.show()
```

```python
pr_files = [
    '/content/drive/MyDrive/ethereum-fraud-detection/results/precision_recall_data_logistic_reg.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/precision_recall_data_svm.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/precision_recall_data_neural_net.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/precision_recall_data_random_forest.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/precision_recall_data_xgboost.csv',
    '/content/drive/MyDrive/ethereum-fraud-detection/results/precision_recall_data_lightgbm.csv',
]

labels = [
    'Logistic Regression',
    'SVM',
    'MLP',
    'Random Forest',
    'XGBoost',
    'LightGBM',
]

colors = ['cyan', 'brown', 'olive', 'magenta', 'blue', 'green']

plt.figure(figsize=(8, 6))

for i, file in enumerate(pr_files):
    pr_data = pd.read_csv(file)
    precision = pr_data['precision']
    recall = pr_data['recall']
    auc_value = auc(recall, precision)

    plt.plot(recall, precision, color=colors[i], label=f'{labels[i]} (AUC={auc_value:.2f})', linewidth=1.0)

plt.title('Precision-Recall (PR) Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='lower left')
# plt.grid(True)
plt.show()
```
